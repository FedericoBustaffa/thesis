```python
import pandas as pd

df = pd.read_csv("results/genetic_regression.csv")
df = df.rename(
    columns={
        "samples": "Samples",
        "numpy_MSE": "Numpy",
        "ppga_MSE": "PPGA",
        "deap_MSE": "DEAP",
    }
)
df
```

```python
print(df.to_latex(index=False, float_format="%.3g"))
```

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams.update({"font.size": 16})

fig = plt.figure(figsize=(6, 4.5), dpi=300)
plt.plot(df["Samples"], df["Numpy"], "gx--", label="Numpy")
plt.plot(df["Samples"], df["DEAP"], "rs-", label="DEAP")
plt.plot(df["Samples"], df["PPGA"], "bo-", label="PPGA")

plt.xlabel("Punti")
plt.ylabel("Mean Square Error")
plt.xscale("log", base=2)
plt.xticks(df["Samples"], df["Samples"])
plt.yticks(
    [300000, 310000, 320000, 330000, 340000, 350000],
    labels=[3, 3.1, 3.2, 3.3, 3.4, 3.5],
)
plt.grid(True)
plt.legend()
plt.tight_layout()
fig.text(0.1, 0.975, "e+05", fontsize="12")
plt.savefig("/home/federico/tesi/immagini/regression_trend.svg")
plt.show()
```
