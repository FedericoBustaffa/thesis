```python
import pandas as pd

greedy_df = pd.read_csv("results/greedy.csv")
deap_df = pd.read_csv("results/deap.csv")
ppga_df = pd.read_csv("results/ppga.csv")

df = greedy_df.rename(
    columns={
        "items": "Oggetti",
        # "capacity": "Capacit√†",
        "value": "Greedy",
        # "weight": "Peso",
    },
    inplace=False,
).drop(columns=["weight", "capacity"])
df["PPGA"] = ppga_df["value"]
df["DEAP"] = deap_df["value"]
df
```

```python
print(df.to_latex(index=False, float_format="%.2f"))
```

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams.update({"font.size": 16})

plt.figure(figsize=(6, 4.5), dpi=300)
plt.plot(df["Oggetti"], df["Greedy"], "gx--", label="Greedy")
plt.plot(df["Oggetti"], df["DEAP"], "rs-", label="DEAP")
plt.plot(df["Oggetti"], df["PPGA"], "bo-", label="PPGA")

plt.xlabel("Numero Oggetti")
plt.ylabel("Valore")
plt.xscale("log", base=2)
plt.xticks(df["Oggetti"], df["Oggetti"])
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("/home/federico/tesi/immagini/knapsack_trend.svg")
plt.show()
```
